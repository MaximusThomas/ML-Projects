```python
def initialize_parameters(layer_dims):
    '''
    E.g. layer_dims = [100, 10, 25, 15, 1]
    '''

    parameters = {}
    for l in range(1, len(layer_dims)):
        parameters['W'+str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1]) * np.sqrt(2/layer_dims[l-1])
        parameters['b'+str(l)] = np.zeros((layer_dims[l], 1))
    
    return parameters
```

```python
def sigmoid(x):
    
    sigmoid = 1/(1+np.exp(-x))
```

```python
def forward_prop(A_in, W, b, activation):

    Z = np.dot(W, A_in) + b
    if activation=='relu':
        A_out = max(0, Z)
    elif activation=='tanh':
        A_out = np.tanh(Z)
    elif activation=='sigmoid':
        A_out = sigmoid(Z)

    z_cache = {'A_in':A_in, 'W':W, 'b':b}
    a_cache = {'Z': Z}
    cache = (z_cache, a_cache)

    return A_out, cache
```

```python
def binary_cross_entropy_cost(Yhat, Y):
    
    m = Y.shape[1]
    cost = -1/m * np.sum((Y * np.log(Yhat))+((1-Y)*np.log(1-Yhat)))
    
    return cost
```

```python
def back_prop(dZ, dA_out, cache, activation):
    '''
    Requires A_in, dA_out, W, and b
    '''
    A_in, W = cache[0]['A_in'], cache[0]['W']
    m = A_in.shape[1]

    if activation=='relu':
        dZ = dA_out * (dA_out > 0)
    elif activation=='tanh':
        dZ = dA_out * (1-np.square(np.tanh(dA_out)))
    elif activation=='sigmoid':
        dZ = dA_out * sigmoid(dA_out) * (1-sigmoid(dA_out))
    
    dW = 1/m * np.dot(dZ, A_in.T)
    db = 1/m * np.sum(dZ)
    dA_in = np.dot(W.T, dZ)

    return dA_in, dW, db
```
